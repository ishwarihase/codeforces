```python
import time # to cehck qhich oppp is faster
size=100000
arr=np.arange(size)
arr
```

```python
resultlist= [0]*size
print(resultlist)
```

```python
# pure python
starttimepy =time.time()
for i in range(size):
    resultlist[i]=arr[i]*2 +5
endtimepy=time.time()
finaltimepy=endtimepy-starttimepy
print(finaltimepy)
```

```python
#numpy vectorization
starttimeNP =time.time()
answer= arr *2 +5
endtimeNP=time.time()
finaltimeNP=endtimeNP-starttimeNP
print(finaltimeNP)
```

```python
speed= finaltimepy / finaltimeNP
print(f'numpy is approximately {speed:.1f}x faster')
```
