```python
# filter data baes on condition
df.head()
# i want to [print values where sepal lenth >6]

df[df['sepal_length'] > 6.0]
```

```python
x=df[(df['petal_length']>6) & (df['petal_length']<7)]
print(x)
```

```python
# how to modify existing data
x['petal_length']= x['petal_length'] + 1
x
```

```python
#create nwe column
df['newcol'] = df['sepal_length'] * 10
df.head()
```

```python
#rename column
df.rename(columns={'newcol': 'updated column'},inplace=True) # inpace true means changes are getting saves
df.head()
```

```python
# to check duplicates values 
df.duplicated()
df.duplicated().sum()
```

```python
#to remove duplicated values
df.drop_duplicates(inplace = True)
df.shape
```
